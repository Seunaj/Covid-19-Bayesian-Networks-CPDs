```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.ExcelFile('CPDs for Death.xlsx')
df1 = pd.read_excel(data, 'No')
df2 = pd.read_excel(data, 'Yes')
x = df1.values
z = df2.values
x1 = ['0-9 Yrs', '10-19 Yrs', '20-29 Yrs', '30-39 Yrs', '40-49 Yrs', '50-59 Yrs', '60-69 Yrs', '70-79 Yrs', '80+ Yrs']

fig=plt.figure()
fig.set_size_inches(19.0, 7.0)

ax=fig.add_subplot(121)
ax.plot(x1,x[0,1:],c='#ff0000',marker="^",ls='-',label='L=No,M=No,O=No',fillstyle='none')
ax.plot(x1,x[1,1:],c='#800000',marker=(8,2,0),ls='-',label='L=No,M=No,O=Yes')
ax.plot(x1,x[2,1:],c='#008000',marker="v",ls='-',label='L=No,M=Yes,O=No')
ax.plot(x1,x[3,1:],c='#0000cd',marker="o",ls='-',label='L=No,M=Yes,O=Yes',fillstyle='none')
ax.plot(x1,x[4,1:],c='#ff0000',marker="+",ls='--',label='L=Yes,M=No,O=No')
ax.plot(x1,x[5,1:],c='#800000',marker="o",ls='--',label='L=Yes,M=No,O=Yes')
ax.plot(x1,x[6,1:],c='#008000',marker="^",ls='--',label='L=Yes,M=Yes,O=No')
ax.plot(x1,x[7,1:],c='#0000cd',marker="v",ls='--',label='L=Yes,M=Yes,O=Yes')
ax.set_xlim(-0.5, 8.5)
ax.set_ylim(0, 1.05)
ax.grid()
plt.legend(loc=3)
plt.title('(a) G = No')
plt.ylabel('Probability')
plt.xlabel('Age groups')


ax=fig.add_subplot(122)
ax.plot(x1,z[0,1:],c='#ff0000',marker="^",ls='-',label='L=No,M=No,O=No',fillstyle='none')
ax.plot(x1,z[1,1:],c='#800000',marker=(8,2,0),ls='-',label='L=No,M=No,O=Yes')
ax.plot(x1,z[2,1:],c='#008000',marker="v",ls='-',label='L=No,M=Yes,O=No')
ax.plot(x1,z[3,1:],c='#0000cd',marker="o",ls='-',label='L=No,M=Yes,O=Yes',fillstyle='none')
ax.plot(x1,z[4,1:],c='#ff0000',marker="+",ls='--',label='L=Yes,M=No,O=No')
ax.plot(x1,z[5,1:],c='#800000',marker="o",ls='--',label='L=Yes,M=No,O=Yes')
ax.plot(x1,z[6,1:],c='#008000',marker="^",ls='--',label='L=Yes,M=Yes,O=No')
ax.plot(x1,z[7,1:],c='#0000cd',marker="v",ls='--',label='L=Yes,M=Yes,O=Yes')
ax.set_xlim(-0.5, 8.5)
ax.set_ylim(0, 1.05)
ax.grid()

plt.legend(loc=2)
plt.title('(b) G = Yes')
plt.ylabel('Probability')
plt.xlabel('Age groups')
#plt.show()
plt.savefig('CPD_Death.pdf', bbox_inches='tight')
```
